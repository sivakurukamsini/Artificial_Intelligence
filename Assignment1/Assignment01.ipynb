```python
import builtins
print = builtins.print

bmi = float(input("Enter the BMI index: "))

if (bmi < 18.5):
    print("Underweight")
elif (18.5 <= bmi <= 24.9):  
    print("Healthy")
elif (25 <= bmi <= 29.9):    
    print("Overweight")
else:
    print("Obese")
    
    
```
